# use cosine similarity to measure how close the 2 playlists are


```python
from sklearn.metrics import pairwise
```

```python
pairwise.cosine_similarity(np.array(avg_audio_values.loc[overlap, :]).reshape(1, -1), np.array(avg_audio_values2.loc[overlap, :]).reshape(1, -1))
```

```python
def evaluate(summarized_values1, summarized_values2):
    overlap = [column for column in summarized_values1.index.values if column in summarized_values2.index.values]
#     print(overlap)
    
    mse = mean_squared_error(summarized_values1.loc[overlap, :], summarized_values2.loc[overlap, :])
    print('RMSE', np.sqrt(mse))
    
    cosine_sim = pairwise.cosine_similarity(np.array(summarized_values1.loc[overlap, :]).reshape(1, -1), np.array(summarized_values2.loc[overlap, :]).reshape(1, -1))[0]
    print('cosine_sim', np.sqrt(cosine_sim))
```

```python
evaluate(avg_audio_values, avg_audio_values2)
```
